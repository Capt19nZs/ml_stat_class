```python
# from sklearn.model_selection import train_test_split
y1 = mnist["target"].astype(int)
y2 = np.where(y1 % 2 == 0,1,0)
X = mnist["data"]
X_train, X_val,  y1_train, y1_val, y2_train, y2_val = train_test_split(X, y1, y2, random_state=42)
```

```python
# ANSWER
np.random.seed(42)
tf.random.set_seed(42)

# use keras.optimizers.SGD(lr=1e-3)
# epochs=20

inputs = keras.layers.Input(shape=[28*28], name="deep_input")
hidden1 = keras.layers.Dense(150, activation="relu")(inputs)
hidden2 = keras.layers.Dense(70, activation="relu")(hidden1)
hidden3 = keras.layers.Dense(30, activation="relu")(hidden2)
out1 = keras.layers.Dense(10, name="main_output")(hidden3)
out2 = keras.layers.Dense(2, name="aux_output")(hidden3)
model = keras.models.Model(inputs=[inputs],
                           # use two outputs in a model
                           outputs=[out1, out2])
#The model tried to minimize weighted sum of the MSE of the two outpus
model.compile(loss="sparse_categorical_crossentropy", metrics=["accuracy"], loss_weights=[0.5, 0.5], optimizer=keras.optimizers.SGD(lr=1e-3))
history = model.fit(X_train, [y1_train, y2_train], epochs=20,
                    validation_data=(X_val, [y1_val, y2_val]))
```
